## 霍夫变换

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

%config InlineBackend.figure_format = 'retina'

```

### 假定在边缘检测时，有如下监测到的点都属于边缘

```python
x = np.random.rand(4)
y = np.arange(1, 5)
points = [[a, b] for (a, b) in zip(x, y)]
count = 360

plt.scatter(x, y)
```

1. 直角坐标系下的每条直线都对应于极坐标系下的一个点$(\theta, \rho)$；
2. 那么对于每一个点，每间隔1度画出一条直线，一共得到360条直线，把这360条直线画到对应到极坐标系下，就是一条新的曲线；
3. 对于每个边缘点，执行相同的操作，得到下图所示的极坐标系下的曲线
那么，对于原来就处于同一直线上的两点，它们必定在各自的360条直线中有一条共同的直线，这条共同直线对应到极坐标系下就是一个点

```python

for (x, y) in points:
    theta = np.arange(0, count, 1)
    rho = x * np.cos(2*np.pi*theta/count) + y * np.sin(2*np.pi*theta/count)
    plt.plot(theta, rho)
```

4. 这些曲线的交点都表示一条原图中监测到的直线。
